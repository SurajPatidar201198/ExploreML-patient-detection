# Patient Conversation Prediction

```python
#Importing Libraries
import pandas as pd
import numpy as np
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import csv
import numpy
from nltk.tokenize import word_tokenize


#getting dummies
uploaded = pd.read_csv("train.csv",encoding = 'unicode_escape')
uploaded['Source'] = uploaded['Source'].str.lower()

dummies = pd.get_dummies(uploaded['Source'])
dummies
```

```python
#getting file
data_1 = open('train.csv', errors='ignore')
raw1 =data_1.read()

#getting file
data_2 = open('test.csv', errors='ignore')
raw2 = data_2.read()
raw = raw1 + raw2

#Tokenizing word
raw_tok = word_tokenize(raw)

#removing extra symbol and letters
punctuations = '''!()[]{};:.,"\<>/?@#$%^&*_~'0123456789` '''
for elem in punctuations:
    raw_tok = [item.replace(elem, "") for item in raw_tok]

#getting lower form of all letters    
raw_lower  = [word.lower() for word in raw_tok]
#filter out whitespaces
raw_lower = list(filter(None, raw_lower))


#removing stopwords
stop_words = set(stopwords.words('english'))
filtered_word = []
for word in raw_lower:
    if word not in stop_words:
        filtered_word.append(word)
        

#pst = PorterStemmer()
#stem_word = [pst.stem(word) for word in filtered_word]


Vocab = nltk.FreqDist(filtered_word)
len(Vocab)
```

```python
del Vocab['nt']
```

```python
def open_with_csv(filename, d='\t'):
    data = []
    with open(filename, errors = 'ignore') as tsvin:
        tie_reader = csv.reader(tsvin, delimiter=d)
        for line in tie_reader:
            data.append(line)
        return data

data_from_csv = open_with_csv('train.csv')

data_token=[]
for sentence in data_from_csv:
    data_token.append([word_tokenize(item) for item in sentence])
```

```python
punctuations = '''!()[]{};:.,"\<>/?@#$%^&*_~'0123456789`'''
tooo=[]
for sentence in data_token:
    ss=[]
    for word in sentence:
        
        for elem in punctuations:
            word = [item.replace(elem,"") for item in word]
        tooo.append(word)
```

```python
lowerlist = lambda x: [ y.lower() for y in x]
loww= list(map(lowerlist, tooo))
```

```python

ver = []
for sentence in loww:
    bag_word = numpy.zeros(len(Vocab))
    for w in sentence:
        for i, word in enumerate(Vocab):
            if word == w:
                bag_word[i] += 1
    ver.append(bag_word)
    
   
```

```python
df = pd.DataFrame(ver)
df.head()
```

```python
#adding dummies Values
df_1 = pd.DataFrame()
df_1['33720'] = dummies['blog']
df_1['33721'] = dummies['facebook']
df_1['33722'] = dummies['forums']
df_1['33723'] = dummies['youtube']
df_1
```

```python
target = pd.read_csv("train.csv",encoding = 'unicode_escape')['Patient_Tag'].tolist()
```

```python
df = df[:-1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix
X_train, X_test, y_train, y_test = train_test_split(df,target,test_size=0.10, random_state = 5)
```

```python
from sklearn.naive_bayes import MultinomialNB

clf = MultinomialNB(alpha=1)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(metrics.accuracy_score(y_test, y_pred))
```

```python
score = confusion_matrix(y_test, y_pred)
score
```
